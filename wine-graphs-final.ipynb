```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing both data
red = pd.read_csv('Resources/Data/winequality-red.csv', delimiter=';')
white = pd. read_csv('Resources/Data/winequality-white.csv', delimiter=';')
```

```python
#combining two datas into one
#create color segment for each data set
color_red = np.repeat('red', red.shape[0])
color_white = np.repeat('white', white.shape[0])

#append colums to each data set
red['color'] = color_red
white['color'] = color_white

#combine two dataframe
all_wine = red.append(white)

all_wine
```

```python
var = red.columns.tolist()
print(var)
```

# Red Wine

```python
red.describe()
```

## Quality vs all factors
#### higher quality wine has higher citric acid, sulphates, alcohol, lower volatile acidity, (lower density, pH)

```python
sns.set_style('ticks')
sns.set_context('notebook')
var = red.columns.tolist()
plt.figure(figsize = (20, 16))
for i in range(12):
    plt.subplot(3,4, i+1)
    sns.boxplot(x = red['quality'], y = var[i], data = red)
    plt.ylabel(var[i], fontsize = 12)
    plt.savefig(f"Images/red_boxplot")
plt.show()
```

```python
plt.figure(figsize = (20,16))
sns.heatmap(red[var].corr(), linewidths=1.0, annot=True, fmt='0.2f', cmap="PuBuGn")
plt.savefig(f"Images/red_heatmap")
plt.show()
```
